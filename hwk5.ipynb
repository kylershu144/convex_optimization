```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

### A6.2) Minimax rational fit to the exponential

```python
k = 201
t = [-3 + 6 * (i-1) / (k-1) for i in range(1, k+1)]
y = [np.exp(t[i]) for i in range(k)]
print(len(t), len(y))
```

```python
eps = 1e-3

a0 = cp.Variable()
a1 = cp.Variable()
a2 = cp.Variable()
b1 = cp.Variable()
b2 = cp.Variable()

u = 1
l = 0
max_iters = np.ceil(np.log2((u-l) / eps)).astype(int)
print(f"max_iters: {max_iters}")
iter = 0
while  u-l > eps and iter <= max_iters:
    iter += 1
    gamma = 0.5 * (u + l)
    
    constraints = []
    for i in range(k):
        p = a0 + a1*t[i] + a2*t[i]**2
        q = 1 + b1*t[i] + b2*t[i]**2

        constraints += [
            q >= eps,
            p - y[i]*q <= gamma * q,
            p - y[i]*q >= -gamma * q
        ]

    prob = cp.Problem(cp.Minimize(0), constraints)
    prob.solve(qcp=True)
    if prob.status == 'optimal':
        u = gamma
        print(f"iter: {iter}, u: {u}")
    else:
        l = gamma
```

```python
gamma = u
print(f"gamma: {gamma}")
a0 = cp.Variable()
a1 = cp.Variable()
a2 = cp.Variable()
b1 = cp.Variable()
b2 = cp.Variable()

constraints = []

for i in range(k):
    p = a0 + a1*t[i] + a2*t[i]**2
    q = 1 + b1*t[i] + b2*t[i]**2

    constraints += [
        q >= eps,
        p <= (y[i] + gamma) * q,
        p >= (y[i] - gamma) * q,
    ]

prob = cp.Problem(cp.Minimize(0), constraints)
prob.solve()
prob.status
```

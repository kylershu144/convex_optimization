```python
import cvxpy as cp
```

```python
x = cp.Variable()
y = cp.Variable()
constr = [2*x+y >=1, x+3*y >=1, x>=0, y>=0]
```

```python
obj1 = cp.Minimize(x + y)
prob1 = cp.Problem(obj1, constr)
prob1.solve()
print("optimal value", prob1.value)
print("optimal var", x.value, y.value)
```

```python
obj2 = cp.Minimize(-x-y)
prob2 = cp.Problem(obj2, constr)
prob2.solve()
print("optimal value", prob2.value)
print("optimal var", x.value, y.value)
```

```python
obj3 = cp.Minimize(x)
prob3 = cp.Problem(obj3, constr)
prob3.solve()
print("optimal value", prob3.value)
print("optimal var", x.value, y.value)
```

```python
obj4 = cp.Minimize(cp.maximum(x, y))
prob4 = cp.Problem(obj4, constr)
prob4.solve()
print("optimal value", prob4.value)
print("optimal var", x.value, y.value)
```

```python
obj5 = cp.Minimize(x**2 + 9*y**2)
prob5 = cp.Problem(obj5, constr)
prob5.solve()
print("optimal value", prob5.value)
print("optimal var", x.value, y.value)
```
